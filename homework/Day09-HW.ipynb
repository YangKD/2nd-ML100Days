```python
import numpy as np
np.random.seed(1)
import pandas as pd
import matplotlib.pyplot as plt 

%matplotlib inline
```

# 相關弱

```python
# 隨機生成兩組 1000 個介於 0~50 的數的整數 x, y, 看看相關矩陣如何

x = np.random.randint(0,50,1000)
y = np.random.randint(0,50,1000)

corr = np.corrcoef(x , y)
print(corr)
```

```python
plt.scatter(x,y)
```

# 正相關

```python
# 隨機生成兩組 1000 個介於 0~50 的數的整數 x, y, 看看相關矩陣如何

x = np.random.randint(0,50,1000)

# 這次讓 y 與 x 正相關，再增加一些雜訊

y = x + np.random.normal(0,10,1000)

corr_p = np.corrcoef(x,y)
print(corr_p)
```

```python
plt.scatter(x , y)
```

# 負相關

```python
# 隨機生成兩組 1000 個介於 0~50 的數的整數 x, y, 看看相關矩陣如何

x = np.random.randint(0,50,1000)

# 這次讓 y 與 x 負相關，再增加一些雜訊

y = -x + np.random.normal(0,10,1000)

corr_n = np.corrcoef(x,y)
print(corr_n)
```

```python
plt.scatter(x,y)
```
