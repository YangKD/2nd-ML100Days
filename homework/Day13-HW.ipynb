```python
import os 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
dir_data = "/Users/steven/Part01"
f_app_train = os.path.join(dir_data , 'application_train.csv')
app_train = pd.read_csv(f_app_train)
app_train.shape
```

```python
app_train.head(5)
```

```python
# 將只有兩種值的類別型欄位, 做 Label Encoder, 計算相關係數時讓這些欄位可以被包含在內

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# 檢查每一個 column

for col in app_train:
    if app_train[col].dtype == "object":
        if len(list(app_train[col].unique())) <= 2:
            app_train[col] = le.fit_transform(app_train[col])
            
print(app_train.shape)
app_train.head()
```

```python
# 受雇日數為異常值的資料, 另外設一個欄位記錄, 並將異常的日數轉成空值 (np.nan)

app_train["DAYS_EMPLOYED_ANOM"] = app_train["DAYS_EMPLOYED"] == 365243

app_train["DAYS_EMPLOYED"].replace({365243: np.nan} , inplace = True)


# 出生日數 (DAYS_BIRTH) 取絕對值 
app_train["DAYS_BIRTH"] = abs(app_train["DAYS_BIRTH"])
```

# 練習時間

```python
app_train["YEARS_BIRTH_CUT"] = pd.cut(app_train["DAYS_BIRTH"] / 365 , 10) 
app_train["YEARS_BIRTH_CUT"].value_counts().sort_index()

import seaborn as sns 
fig = plt.figure(figsize = (10,10))
ax = sns.barplot(app_train["YEARS_BIRTH_CUT"] , app_train["TARGET"])
plt.xticks(rotation = 60)
ax.set_title("YEARS_BIRTH_DISCRET By TARGET")
```

```python
app_train["YEARS_EMPLOYED_CUT"] = pd.cut(app_train["DAYS_EMPLOYED"] / 365 , 10)
app_train["YEARS_EMPLOYED_CUT"].value_counts().sort_index()

fig = plt.figure(figsize = (10,10))
ax = sns.barplot(app_train["YEARS_EMPLOYED_CUT"] , app_train["TARGET"])
plt.xticks(rotation = 70)
ax.set_title("YEARS_EMPLOYED_DISCRET By TARGET")
```
